## Scaling the Data

scaling : Fitting our data to the same ranges. 

** Whether a feature is important or not in ML is expressed with coefficients. The ranges of my features can have different range values for each. If we don't do a scale, features in a large range are weighted more (valid for distance-based models) and our model will return incorrect results. We use scale to prevent this. 

**It also accelerates the work of Gradient descent-based models such as linear, logistic regression.

**There is no disadvantage of applying scale to data. data scale If you are confused about whether to apply it or not, you can apply scale directly.

Data leakage: Leakage of information about the data in the test set to the data in the train set

** To prevent leakage, only fit operation is applied to X_train. On the other hand, only transform operation is applied to X_test. no fit operation is applied to all data.

** Target is never scaled. zaten bir sütun başka bir sütunla karışma karşılaştıma yok targetlar arası karşılaştırma diye bir şey yok.

scaling bir data preprocessing işlemidir,
modele sokmadan önce yapılan bir işlem

feature lar daki her brindeki bir birimlik değişim birbirine eşit değil. Skale ling ile bu eşit  oluyor.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler # Robustscaler is used when outlier could be present

scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # Apply just for X_tarin not for X_test (Kalıp çıkarma)


# x_traini fit ettik dikkat X'i değil.burada kalıbımı çıkarıyor.
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled                                
```

## Polinomial Regression Model Building

### linear regresyona Polinomial featureları koyarak fit edersek bu --------> Polinomial Regression olur.

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()   # model kuruldu
```

```python
lm.fit(X_train_scaled, y_train)   # model eğitildi ve scale edilmiş x_train'i aldığımıza dikkat edelim.
```

```python
##  intersept ve coefler bulundu
```

```python
y_pred = lm.predict(X_test_scaled)
y_train_pred = lm.predict(X_train_scaled)     ## bunu yapmamızın nedeni overfitting-underfitting var mı bakacağız.
```

```python
## hem traini hem testi göstersin bana bir def ile : karşılaştırmak için.
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, name):
    
    scores = {name+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    name+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear") # Evaluate the result. Overfitting?
ls
```

#### traindeki R2 teste göre çok yüksek ve hata metricleri de bununla doğru orantılı olduğundan overfitting diyebiliyoruz rahatlıkla

## test ve train datasının eror metrikleri arasındaki farka varyans diyoruz

## <span style='color:red'> Multicolineratiy</span>  featurelar arasındaki çok güçlü korelasyon

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
def color_red(val):
    if abs(val) > 0.90 and abs(val) < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'
```

```python
## abs(val)'ı kullanabiliriz.
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)
```

```python
## burada bu kadar olmasının nedeni -------> polinomial featureslardan dolayı
```

### multicolineartiy linear regresyonda bazı sıkıntılara neden olabilir : 

----> modelin unstable'a itiyor

## <span style='color:red'> Cross Validation</span> 

We do cross-validation to check whether the one-time scores we receive are consistent or not

cross validation is only applied to the train set.

```python
from sklearn.metrics import SCORERS
list(SCORERS.keys())
```

```python
from sklearn.model_selection import cross_validate
```

```python
model = LinearRegression()
scores = cross_validate(model, X_train_scaled, y_train, scoring = ['r2', 'neg_mean_absolute_error','neg_mean_squared_error', 
                                                            'neg_root_mean_squared_error'], cv = 5)
```

                    crossvalidate içine ne verirseniz onu cross validat eder

#### scoring kısmı önemli dikkat edelim.cross_validate algoritmasının hepsini birden maksimize etmesi istediğinden dolayı error metricleri negatif olarak yazılıyor.

## <span style='color:red'> Grid Search</span> 

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42)
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_
```
