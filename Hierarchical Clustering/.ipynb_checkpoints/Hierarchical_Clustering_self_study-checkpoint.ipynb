```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
X =df.drop("species",axis=1)
```

```python
X
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

### linkage()


## dendrogram()





## Z : Linkage'Ä± temsil ediyor dikkat

```python
hie_complete = linkage(X,method="complete")
hie_ward = linkage(X,method="ward")
```

```python
dendrogram(hie_ward);
```

```python
dendrogram(hie_complete);
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc_model = AgglomerativeClustering(n_clusters=3,linkage="ward")
hc_model.fit_predict(X)
```

```python
hc_model.labels_
```

```python
clusters = hc_model.labels_
```

```python
df["predicted_clusters"] = clusters
```

```python
df
```

```python
pd.crosstab(df["predicted_clusters"],df["species"])
```

```python
df["species_"] =  df["species"].map({"setosa":1,"versicolor":0,"virginica":2})
```
