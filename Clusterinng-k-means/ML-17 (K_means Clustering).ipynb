# Choosing The Optimal Number of Clusters

## Elbow metod

```python
ssd = []                                     #sum squared distance

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)   
```

  # intertia: Distance from each sample to centroid of its cluster (SSD)

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia/error (ssd)") 
plt.title("elbow method") 
```

## 3 m seeceiz 4 m buna karar veremedik diyelim farkl eylere de bakabiliriz.

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff()   # dler
```

## Silhouette analysis

    ##her bir nokta iin kme ii mesafe hesaplanp (kendi merkezlerine olan uzaklk) ORTALMASI ALINIYOR ---------------> a
    her bir nokta iin  bir sonraki en yakn kmedeki dier tm noktalar arasndaki  ORTALMASI ALINIYOR -----------> b
    

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)           # k ------ 5 idi burada hatrlarsak
```

# k deiseydi silhouette_score ne olurdu ? 

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

## ------> kararsz kaldmz nokta 3 m 4 m olsun idi yukarda o yzden 3 ve 4'te hangisinde silhouette_score yksekse ona odaklanmamz gerekir.

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)                     # 3'e karar verdik inceleyelim imdi.
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```
