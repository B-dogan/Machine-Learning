# Choosing The Optimal Number of Clusters

## Elbow metod

```python
ssd = []                                     #sum squared distance

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)   
```

  # intertia: Distance from each sample to centroid of its cluster (SSD)

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia/error (ssd)") 
plt.title("elbow method") 
```

## 3 mü seçeceğiz 4 mü buna karar veremedik diyelim farklı şeylere de bakabiliriz.

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff()   # düşüşler
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

## Silhouette analysis

    ##her bir nokta için küme içi mesafe hesaplanıp (kendi merkezlerine olan uzaklık) ORTALMASI ALINIYOR ---------------> a
    her bir nokta için  bir sonraki en yakın kümedeki diğer tüm noktalar arasındaki  ORTALMASI ALINIYOR -----------> b
    

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)           # k ------ 5 idi burada hatırlarsak
```

# k değişseydi silhouette_score ne olurdu ? 

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

## ------> kararsız kaldığımız nokta 3 mü 4 mü olsun idi yukarıda o yüzden 3 ve 4'te hangisinde silhouette_score yüksekse ona odaklanmamız gerekir.

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)                     # 3'e karar verdik inceleyelim şimdi.
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```
